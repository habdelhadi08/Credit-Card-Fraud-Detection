```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('dataset\creditcard.csv')
df
```

```python
def summarize_dataset(df):
    print("Shape of the dataset:")
    print(df.shape)
    
    print("\nDescriptive statistics of the dataset:")
    print(df.describe())
    
    print("\nDataset info (including null values and data types):")
    df.info()  
    
    print("\nMissing values in each column:")
    print(df.isnull().sum())
    
    print("\nNumber of duplicate rows:")
    print(df.duplicated().sum())
    
   

    
    
    print("\nFirst 20 rows of the dataset:")
    return df.head(20)




summarize_dataset(df)




    
```

```python
# Drop duplicate rows
df = df.drop_duplicates()

# Confirm removal
print("Number of duplicate rows after removal:", df.duplicated().sum())
print("New dataset shape:", df.shape)

# Save the cleaned DataFrame to a new CSV file
df.to_csv('cleaned_creditcard.csv', index=False)
print("Cleaned dataset saved as 'cleaned_creditcard.csv'")
```

```python
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
import numpy as np

def visualize_dataset(df):
    # Visualize class distribution using Plotly
    fig = px.bar(df['Class'].value_counts().reset_index(), x='index', y='Class', 
                 labels={'index': 'Class', 'Class': 'Count'}, 
                 title='Class Distribution (0 = Non-Fraud, 1 = Fraud)')
    pio.show(fig)  # Use plotly.io.show() instead of fig.show()



    # Visualize transaction amount distribution by fraud status using Plotly
    fig = px.histogram(df, x='Amount', color='Class', 
                        nbins=50, 
                        title='Transaction Amount Distribution by Fraud Status', 
                        labels={'Amount': 'Transaction Amount', 'Class': 'Fraud Status'})
    pio.show(fig)  # Use plotly.io.show() instead of fig.show()

# Run the visualization function
visualize_dataset(df)



```
